```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ucop.edu/operating-budget/budgets-and-reports/legislative-reports/2017-18-legislative-session.html'
```

```python
result = requests.get(url)
c = result.content
soup = BeautifulSoup(c)
```

```python
summary = soup.find('table',{'class':'table-striped', 'id':'report'})
summary
```

```python
tables = summary.find_all('tbody')
```

```python
tables
```

```python
data = []
rows = tables[0].findAll('tr')
for tr in rows:
    cols = tr.findAll('td')
    for td in cols:
        text = td.find(text = True)
        print(text)
        data.append(text)
```

```python
data
```

```python
data1 = []
for item in data:
    if item != None:
        data1.append(item)
data1
```

```python
reports = []
date = []
index = 0
for item in data1:
    if 'pdf' in item:
        date.append(data1[index-1])
        reports.append(item)
    index += 1
```

```python
reports
```

```python
date
```

```python
date = pd.Series(date)
```

```python
reports = pd.Series(reports)
```

```python
df = pd.concat([date, reports], axis = 1)
```

```python
df.column = ['Dates','Reports']
df
```
